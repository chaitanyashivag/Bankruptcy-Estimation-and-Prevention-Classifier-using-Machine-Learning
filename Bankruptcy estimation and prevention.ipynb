# 1. Logistic Regression Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
```

```python
x=df.iloc[:,0:6]
y=df.iloc[:,-1]
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.3,random_state=4)
```

```python
x_train
```

```python
model1=LogisticRegression()
model1.fit(x_train,y_train)
train_pred=model1.predict(x_train)
```

```python
confusion_matrix1 = confusion_matrix(y_train,train_pred)
print (confusion_matrix1)
```

```python
print(classification_report(y_train,train_pred))
```

```python
test_pred=model1.predict(x_test)
```

```python
confusion_matrix2 = confusion_matrix(y_test,test_pred)
print (confusion_matrix2)
```

```python
print(classification_report(y_test,test_pred))
```

# 3. K Nearest Neighbours Classifier

# 5. Decision Tree Classifier

```python
from sklearn.tree import  DecisionTreeClassifier
from sklearn import tree
```

```python
model6 = DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
model6.fit(x_train,y_train) 
```

```python
tree.plot_tree(model6);
```

```python
fn=['industrial_risk','management_risk','financial_flexibility','credibility', 'competitiveness', 'operating_risk']
cn=['non bankruptcy', 'bankruptcy']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(model6,
               feature_names = fn, 
               class_names=cn,
               filled = True); 
```
